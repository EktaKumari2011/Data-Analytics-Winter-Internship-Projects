# **1. Missing Values in DataSet**

Question: Are there any missing values in the dataset, and if so, how should they be handled for each indicator?

Answer:There are 2 columns having missing values Age and Ratings.

-->Age: We will be taking mean value of column and will replace the null value to mean value.

-->Rating: We will be replacing the null values with mean value of column.

```python
#Mean value replacement for age column
df.isnull().sum()
avg_age = df['Age'].mean()
avg_age
df['Age'] = df.Age.round(decimals=0)
df['Age'] = df.Age.fillna(avg_age)
df
#mean value replacement for rating column
avg_rating = df['Rating'].mean()
avg_rating
df['Rating'] = df.Rating.round(decimals=0)
df['Rating'] = df.Rating.fillna(avg_rating)
df
```

# **3. Outliers:**
Question: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?

Answers:

-->printing of data with outliers

-->as the data set is really we will be adjusting the data outliers

Conclusion: The data ouliers are only in AGE and Ratings column they can be adjusted according to the need of analysis



```python
df.Rating.describe()
Q1 = df.Rating.quantile(0.25)
Q3 = df.Rating.quantile(0.75)
IQR = Q3-Q1
print("IQR",IQR)
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
lower_bound
upper_bound
df.loc[(df.Rating<lower_bound) | (df.Rating>upper_bound),'Rating']=np.nan
df['Rating']=df['Rating'].fillna(df.Rating)
df.Rating.plot(kind='box')
df
```

# **4. Salary Formatting:**
Question: Examine the format of the Salary column. Does it require any formatting or standardization for consistent analysis?

Answer:
The format of the salary column is mixed with strings and numbers

--> Formatting can be done like:

a) removing all the symbols

b)Splitting the salary value into two Min_salary and Max_salary

c)Applying conversion of k to Min and Max with the initialisation of numeric data type

--> for code take refrence of code followed by markdown 2

```python
df.Salary = df.Salary.str.replace("$"," ")
df.Salary = df.Salary.str.replace("k","000")
df.Salary = df.Salary.str.replace("-"," -")
df
print(df.dtypes)
```

# **5. Location Standardization:**
Question: Check the consistency of location entries. Do they need standardization, and how can this be achieved?

Answer:
Yes, the location need a standarisation because it is inconsistent. We will be using an abbreviation mapping for cleaning the location column.

```python

df.Location = df.Location.str.replace(',In', ' ')
df.Location = df.Location.str.replace(' In', ' ')
df.Location = df.Location.str.replace('Australia Aus', 'Australia ')
df.Location = df.Location.str.replace(',Ny', ' ')
df


```

# **6. Established Column:**
Question: Explore the Established column. Are there any inconsistencies or anomalies that need to be addressed?

Answer:
Established column had some value entries of (-1) which was replaced by (NA).

```python
#Cleaning Established column
df.loc[df['Established']<=0,'Established']=np.nan

Year = df['Established'].median()
Year
df['Established']=df['Established'].fillna(Year).astype(int)
df
```

# **15. Consistent Rating Scale:**
Question: Ensure a consistent rating scale in the Rating column. Should it be normalized or adjusted for uniform analysis?

Answer:
As earlier we took average rating value to replace the null values so here adjusting the rating column would seem more appropriate method.



```python
#Adjusting the rating column
average_rating = df['Rating'].mean()
target_average = 5.0

df['Adjusted_Rating'] = df['Rating'] - (average_rating - target_average)
df
```
